Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

I used the `schools.csv` file for my analysis that I found on datacamp, which is previewed below.

I have attempted to use my learnings about pandas and pivot tables to find answers to a few questions: 
1. What schools in NYC have the best math results?
2. What are the top 10 performing schools based on the combined SAT scores?
3. Which single borough has the largest standard deviation in the combined SAT score?

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

1. What schools in NYC have the best math results?

```python
# Filtering the school dataframe
best_math = schools[schools["average_math"] >= (0.8 * 800)]

# Taking relevant columns and sorting
best_math_schools = best_math[["school_name", "average_math"]].sort_values("average_math", ascending = False)

# Viewing the dataframe
best_math_schools
```

There are 10 schools in NYC that have had an average math score in the SAT that is higher than 640/800 i.e., atleast 80% of total possible math score.

2. What are the top 10 performing schools based on the combined SAT scores?

```python
# Making the total_SAT column
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# Sorting to find top scoring schools and only required columns
schools_sorted = schools.sort_values("total_SAT", ascending = False)[["school_name", "total_SAT"]]

# Taking a list of only top 10 schools
top_10_schools = schools_sorted[0:10]

# Previewing the dataframe
top_10_schools
```

These are the top 10 schools in NYC with the highest average SAT scores, ranging from 2144 - 1889, i.e., 89.33% - 78.70%

3. Which single borough has the largest standard deviation in the combined SAT score?

```python
# Grouping with respect to borough 
borough_data = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

# rename columns 
borough_data = borough_data.rename(columns={"count" : "num_schools", "mean": "average_SAT", "std": "std_SAT"})

# Displaying data of all 5 borough - 375 schools combined
borough_data
```

```python
# Selecting the borough with the largest std deviation
largest_std_dev = borough_data[borough_data["std_SAT"] == borough_data["std_SAT"].max()]

# Displaying the data of the borough with the largest std dev
largest_std_dev.reset_index(inplace=True)
largest_std_dev
```

The Borough with the largest std dev in SAT scores is thus Manhattan with a standard deviation of 230.29.
